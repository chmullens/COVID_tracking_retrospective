```python
#BUILD THE CORE DATA SET

#    Population (x)
#    Total cases (y)
#    Cases in the last month (s)
#    Cases in the last week (c)

# Size, including all datapoints:
date_start = sdata_rev['date_proper'].min() - dt.timedelta(days=10)
date_end = sdata_rev['date_proper'].max()
days_total = date_end - date_start
array_len = days_total.days + 1

statelist = base_state_stats.sort_values('Estimated pop, 7/1/2019', ascending=False)
#We probably want big states plotted first so the small state dots show up on top
statelist = statelist[['State','USPS abbr','Estimated pop, 7/1/2019',]]
statelist = statelist[statelist['USPS abbr'].agg(lambda x: not set([x]).intersection(small_territory_set))]

# Array size when including all date datapoints:
date_start = sdata_rev['days_since_Jan_22_2020'].min() - 10
date_end = sdata_rev['days_since_Jan_22_2020'].max()
days_total = date_end - date_start + 1

#Assign the animation variables:
dot_dat_x = np.zeros([52, days_total])
dot_dat_y = np.zeros([52, days_total])
dot_dat_s = np.zeros([52, days_total])
dot_dat_c = np.zeros([52, days_total])

dot_dat_death = np.zeros([52, days_total])
dot_dat_testdiff = np.zeros([52, days_total])

tempind = 0

temp_states = []

for state2 in statelist['USPS abbr']:
    temp_states.append(state2)
    
    temp_pop = statelist[statelist['USPS abbr'] == state2]['Estimated pop, 7/1/2019']
    dot_dat_x[[tempind,]] = np.tile(temp_pop,(1,days_total))
    
    sdata_rev_sub = sdata_rev[sdata_rev['state'] == state2]
    sdata_rev_sub = sdata_rev_sub.sort_values('days_since_Jan_22_2020', ascending=True)
    
    postemp = sdata_rev_sub['positive'].fillna(0)
    #Right here is where I have to put a zero for the first diff day. Reported
    #cases prior to reporting are effectively zero. 
    temp = postemp.diff()
    temp.iloc[0] = 0
    sdata_rev_sub['positive_diff'] = temp
    #Tack a month of zeros onto the front end to make rolling avg work better:
    positive_diff_ext = pd.concat([pd.Series(np.zeros([28])),temp])
    temp2 = positive_diff_ext.rolling(window=28).mean()
    temp3 = positive_diff_ext.rolling(window=7).mean()
    sdata_rev_sub['positive_roll28'] = temp2[27:]
    sdata_rev_sub['positive_roll7'] = temp3[27:]
    
    #And deaths:
    temp = sdata_rev_sub['death'].diff()
    temp.iloc[0] = 0
    sdata_rev_sub['death_diff'] = temp
    #Tack a month of zeros onto the front end to make rolling avg work better:
    death_diff_ext = pd.concat([pd.Series(np.zeros([28])),temp])
    temp2 = death_diff_ext.rolling(window=28).mean()
    temp3 = death_diff_ext.rolling(window=7).mean()
    sdata_rev_sub['deaths_roll28'] = temp2[27:]
    sdata_rev_sub['deaths_roll7'] = temp3[27:]
    
    tempmin = sdata_rev_sub['days_since_Jan_22_2020'].min()
    dot_dat_y[tempind, (tempmin-date_start):] = postemp
    dot_dat_s[tempind, (tempmin-date_start):] = sdata_rev_sub['positive_roll28']
    dot_dat_c[tempind, (tempmin-date_start):] = sdata_rev_sub['positive_roll7']
    dot_dat_death[tempind, (tempmin-date_start):] = sdata_rev_sub['deaths_roll7']
    
    dot_dat_testdiff[tempind, (tempmin-date_start):] = sdata_rev_sub['positive_diff']
    #i.e. if min day is 22, day 22 is index zero, days 55-59 are indices 33-37.
    
    tempind += 1
```
